2. Connect + Run Query

```python
from sqlalchemy import create_engine
import pandas as pd

# Connection Info
username = ''
password = ''
host = ''
port = ''
database = ''

# 连接到 mysql 系统库
engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}",
    connect_args={"charset": "utf8mb4"}
)

# ⭐️你的 Query 放在这里, 暂时是一个 placeholder⭐️
query = """
SELECT *
from airlines
"""

# 用 pandas.read_sql
df = pd.read_sql(query, con=engine)

# ⭐️head 就是最多显示多少行, 现在是最多显示 100 行⭐️
df.head(100)
```
